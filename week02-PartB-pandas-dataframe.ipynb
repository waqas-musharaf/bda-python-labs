```python
#pandas untilites and methods are from the pandas package
import pandas as pd
```

pandas has number methods for importing data from files and data stores, e.g. json, excel, csv, html, HDFS

```python
#pandas has a number helper methods for import data from files, e.g. csv, json
#Here we import data from a csv file
adult_data = pd.read_csv(r"C:\Users\Waqas\adult.data")
print(adult_data);
```

Instances from a datafrome can be selected using index slicing.
The format of slicing is [start_index : end_index]. This produces indices from start_index to end_index - 1.

```python
#data can be selected using index slicing
#e.g. the following select the first 3 instances, i.e. 0, 1, 2
adult_data[0:3]
```

```python
adult_data['age']
```

We can also use helper method head and tail to select instances from the top or the bottom of a dataframe.

```python
#Select the first 10 instances from a dataframe
adult_data.head(10)
```

```python
#Select the last 5 instances from a data frame
adult_data.tail(5)
```

While standard list type indexing works, pandas recommends the use of loc and iloc methods for more efficient data selection.

c.f `loc` method https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html#pandas.DataFrame.loc

`iloc` method https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html#pandas.DataFrame.iloc

`iat` method https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iat.html#pandas.DataFrame.iat

```python
#We can use key-label to select specific attributes(columns) from a dataframe
adult_data.loc[0:9,['age','education','salary']]
```

```python
#We can use integer indices rather than key-label to select data from a dataframe
adult_data.iloc[0:9,[0,3,14]]
```

```python
#Use iat to get the 11th instance's age
adult_data.iat[10,0]
```

Conveniently, pandas also allows us to select instances from a dataframe by relational expressions.

```python
#select instances where age is below 40
adult_data[adult_data['age'] < 40]
```

It is often useful to find the unique values of an attribute (column).
This can be obtained by the unique method of a dataframe column.

```python
#obtain the list of unique values from the 'education' column
adult_data['education'].unique()
```

__Task 3:__ Calculate the mean, max and min value of the age and the hours-per-week columns. Please note, for the purpose of this exercise, you are required to write your own calculation instead of using helper functions. 

Hint: In order to do this calcualtion without helper methods, you need to use loop to enumerate the values.

```python
#declare and initialise calculation variables
total_age = 0
max_age = 0
min_age = 999
total_hrs = 0
max_hrs = 0
min_hrs = 999

#iterate through each entry in adult_data
for i in range(len(adult_data)):
    #calculate the sum of the age and hours-per-week values per entry
    total_age = total_age + adult_data.iat[i,0]
    total_hrs = total_hrs + adult_data.iat[i,12]
    #check if the current entry is the smallest or largest of its type discovered so far,
    #and if so, assign it to the appropriate variable
    if (adult_data.iat[i,0] > max_age):
        max_age = adult_data.iat[i,0]
    elif (adult_data.iat[i,0] < min_age):
        min_age = adult_data.iat[i,0]
    if (adult_data.iat[i,12] > max_hrs):
        max_hrs = adult_data.iat[i,12]
    elif (adult_data.iat[i,12] < min_hrs):
        min_hrs = adult_data.iat[i,12]

#print the stored min/max variables, calculate the mean variables by dividing
#the 'total' variables by the number of entries in adult_data
print ("Mean age:",total_age/len(adult_data))
print ("Max age:",max_age)
print ("Min age:",min_age)
print ("Mean hours-per-week:",total_hrs/len(adult_data))
print ("Max hours-per-week:",max_hrs)
print ("Min hours-per-week:",min_hrs)
```
